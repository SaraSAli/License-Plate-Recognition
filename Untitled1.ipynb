```python
'''
What is Histogram Equalization?
Histogram Equalization is an image processing technique that adjusts the contrast of an image by using its histogram.
To enhance the image’s contrast, it spreads out the most frequent pixel intensity values or stretches out the intensity range of the image. 
By accomplishing this, histogram equalization allows the image’s areas with lower contrast to gain a higher contrast.
'''
```

```python
flat_img1 = img1.flatten()
def get_histogram(image, bins):
    # array with size of bins, set to zeros
    histogram = np.zeros(bins)

    # loop through pixels and sum up counts of pixels
    for pixel in image:
        histogram[pixel] += 1

    # return our final result
    return histogram
hist_img1 = get_histogram(flat_img1, 256)
```

```python
plt.plot(hist_img1)
```

```python
hist_img1_b = get_histogram(b, 256)
```

```python
plt.plot(hist_img1_b)
```

```python
hist_img1_g = get_histogram(g, 256)
plt.plot(hist_img1_g)
```

```python
hist_img1_r = get_histogram(r, 256)
plt.plot(hist_img1_r)
```

```python
# create cumulative sum function
def cumsum(a):
    a = iter(a)
    b = [next(a)]
    for i in a:
        b.append(b[-1] + i)
    return np.array(b)

# numerator & denomenator
def num_den(cs):
  nj = (cs - cs.min()) * 255
  N = cs.max() - cs.min()
  return nj / N
```

```python
cs_img1_r = cumsum(hist_img1_r)
cs_img1_g = cumsum(hist_img1_g)
cs_img1_b = cumsum(hist_img1_b)
```

```python
plt.plot(cs_img1_r)
```

```python
plt.plot(cs_img1_g)
```

```python
plt.plot(cs_img1_b)
```

```python
cs_img1_r = num_den(cs_img1_r)
cs_img1_g = num_den(cs_img1_g)
cs_img1_b = num_den(cs_img1_b)
```

```python
plt.plot(cs_img1_r)
```

```python
img1_r_new = cs_img1_r[r]

# put array back into original shape since we flattened it
img1_r_new = np.reshape(img1_r_new, r.shape)

# set up side-by-side image display
fig = plt.figure()
fig.set_figheight(15)
fig.set_figwidth(15)

fig.add_subplot(1,2,1)
plt.imshow(r, cmap='gray')

# display the new image
fig.add_subplot(1,2,2)
plt.imshow(img1_r_new, cmap='gray')

plt.show(block=True)
```

```python
img1_b_new = cs_img1_r[b]

# put array back into original shape since we flattened it
img1_b_new = np.reshape(img1_b_new, b.shape)

# set up side-by-side image display
fig = plt.figure()
fig.set_figheight(15)
fig.set_figwidth(15)

fig.add_subplot(1,2,1)
plt.imshow(b, cmap='gray')

# display the new image
fig.add_subplot(1,2,2)
plt.imshow(img1_b_new, cmap='gray')

plt.show(block=True)
```

```python
image_merge = cv2.merge([img1_r_new, img1_g_new, img1_b_new])
cv2_imshow(image_merge)
```

```python
res = np.hstack((img1, image_merge))
cv2_imshow(res)
```

```python
# Using opencv
equ_b = cv2.equalizeHist(b)
equ_g = cv2.equalizeHist(g)
equ_r = cv2.equalizeHist(r)
```

```python
image_merge_cv = cv2.merge([equ_r, equ_g, equ_b])
```

```python
res_cv = np.hstack((img1, image_merge_cv))
cv2_imshow(res_cv)
```
